```python
#normalisation (same scale)
x = (x - x.min())/(x.max() - x.min())
y = (y - y.min())/(y.max() - y.min())
z = (z - z.min())/(z.max() - z.min())
```

```python
plt.plot(x, y,'r.', z, y,'b.')
```

```python
#converting into a numpy array- faster computation
X = np.array(x)
Y = np.array(y)
Z = np.array(z)
```

```python
#dividing the dataset for training and testing
X_train, X_test, Z_train, Z_test, Y_train, Y_test,  = train_test_split(X, Y, Z, test_size = 0.2)
```

```python
#building functions     
def hypothesis(a, X, Z, b, c):                  #representation
    return a * X + b * Z + c
```

```python
def error(a, X, b, Z, c, Y):                      #evaluation
    e = 0 
    m = len(X)   #total number of features
    
    for i in range(m):
        e += np.power((hypothesis(a, X[i], b, Z[i], c) - Y[i]),2)
        
    return (1 / (2 * m)) *e    
```

```python
def step_gradient(a, X, b, Z, c, Y, learning_rate):                  #optimisation
    grad_a = 0
    grad_b = 0
    grad_c = 0
    m = len(X)
    
    for i in range(m):
        grad_a += (1 / m) * (hypothesis(a, X[i], b, Z[i], c) - Y[i]) * X[i]           #partial differentiation
        grad_b += (1 / m) * (hypothesis(a, X[i], b, Z[i], c) - Y[i]) * Z[i] 
        grad_c += (1 / m) * (hypothesis(a, X[i], b, Z[i], c) - Y[i])
        
    a = a - grad_a * learning_rate
    b = b - grad_b * learning_rate
    c = c - grad_c * learning_rate
    
    return a, b, c
```

```python
def descend(init_a, X, init_b, Z, init_c, Y, learning_rate, iterations):    #iterations- step size to take
    a = init_a     #initializing the starting points
    b = init_b
    c = init_c
    
    for i in range(iterations):
        e = error(a, X, b, Z, c, Y)
        
        if i % 1000 == 0:           # printing error for 1000 steps   
            print(f"Error: {np.sqrt(e)}, a: {a}, b: {b}, c: {c}")
            
        a, b, c = step_gradient(a, X, b, Z, c, Y, learning_rate)
        
    return a, b, c    
```

```python
a = 0
b = 0
c = 0
learning_rate = 0.001
iterations = 20000   #change the value(steps) to reach the minima
```

```python
final_a, final_b, final_c = descend(a, X_train, b, Z_train, c, Y_train, learning_rate, iterations)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.view_init(25,15)
ax.plot(X_train, Z_train, Y_train, "b.")
ax.plot(X_train, Z_train, hypothesis(final_a, X_train, final_b, Z_train, final_c), "g.")
```
