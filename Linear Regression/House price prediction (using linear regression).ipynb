```python
#converting into a numpy array- faster computation
X = np.array(x)
Y = np.array(y)
```

```python
#dividing the dataset for training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
#building functions
def hypothesis(a, X, b):
    return a * X + b
```

```python
def error(a, X, b, Y):
    e = 0
    m = len(X)   #total number of features
    
    for i in range(m):
        e += np.power((hypothesis(a, X[i], b) - Y[i]),2)
        
    return (1 / (2 * m)) *e    
```

```python
def step_gradient(a, X, b, Y, learning_rate):
    grad_a = 0
    grad_b = 0
    m = len(X)
    
    for i in range(m):
        grad_a += (1 / m) * (hypothesis(a, X[i], b) - Y[i]) * X[i]      
        grad_b += (1 / m) * (hypothesis(a, X[i], b) - Y[i])      
        
    a = a - grad_a * learning_rate
    b = b - grad_b * learning_rate
    
    return a, b
```

```python
def descend(init_a, X, init_b, Y, learning_rate, iterations):    #iterations- step size to take
    a = init_a     #initializing the starting points
    b = init_b
    
    for i in range(iterations):
        e = error(a, X, b, Y)
        
        if i % 1000 == 0:           # printing error for 1000 steps   
            print(f"Error: {np.sqrt(e)}, a: {a}, b: {b}")
            
        a, b = step_gradient(a, X, b, Y, learning_rate)
        
    return a,b    
```

```python
a = -0.6308966563287645
b = 0.5766546369867467
learning_rate = 0.001
iterations = 10000   #change the value(steps) to reach the minima
```

```python
final_a, final_b = descend(a, X_train, b, Y_train, learning_rate, iterations)
```

```python
plt.plot(X_train, Y_train, 'r.',
        X_train, hypothesis(a, X_train, b), 'b',
        X_train, hypothesis(final_a, X_train, final_b), 'g')
```
